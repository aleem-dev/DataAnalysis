# Series Plot
Series plot displays value against time

In this example we will import data from yahoo finance library, yfinance.  By far we have loaded data from csv and excel.

```python
import yfinance as yf
import matplotlib.pyplot as plt

# load data, define stock symbol and time range
stock_symbol    =   "AAPL" # apple inc
start_date  =   "2022-01-01"
end_date    =   "2023-01-01"

stock_data = yf.download(stock_symbol,start_date,end_date) # download data

#plot series
plt.figure(figsize=(15,8))
plt.plot(stock_data['Close'], label=f"{stock_symbol} Closing Prices", color='blue')

plt.title(f"{stock_symbol} Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

#download data
stock_symbols = ["AAPL", "INTL", "MSFT", "ASML"]
start_date = "2022-01-01"
end_date = "2023-01-01"

stock_data = yf.download(stock_symbols, start_date, end_date)
    
#plot series, using loop as we have 4 symbol, for in loop
plt.figure(figsize=(15,8))

for sysmbol in stock_symbols:
    plt.plot(stock_data['Close'][sysmbol], label=f"{sysmbol} Closing Prices")

# Alternet approach -> we can only download data for 'Close' and then do looping
# stock_data = yf.download(stock_symbols, start=start_date, end=end_date)['Close']
# for symbol in stock_symbols:
#     plt.plot(stock_data[symbol], label=f"{symbol} Closing Prices")

plt.title("Closing Prices Over Time for Selected Stocks")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.grid()
plt.show()
```
