# Bubble chart
Bubble chart is used to display three variables of data at a time.

In this example we will create a bubble chart to display the height, weight, and the age of a class.

For this example, we will generate random data for the class as shown in the code.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generate dataset for 20 students
np.random.seed(42) # set seed for reproducibility

student_data = {
    'Name': [f'Student{i}' for i in range(1,21)],
    'Height': np.random.uniform(150, 190, 20), # height in centimeter
    'Weight': np.random.uniform(45,90,20), # weight in kilograms
    'Age': np.random.randint(18,25,20), # ages in years
}

student_df = pd.DataFrame(student_data)
student_df.head()

# Create bubble chart
plt.figure(figsize=(12,8))

# Adjust scalling factor for bubble chart
scaling_factor = 500

scatter = plt.scatter(student_df['Height'],student_df['Weight'], s=student_df['Age']*scaling_factor, c=student_df['Age'],cmap='viridis',alpha=0.7)

#add student name to each point label
for i, name in enumerate(student_df['Name']):
    plt.text(
        student_df['Height'][i],
        student_df['Weight'][i],
        name,
        ha='center',
        va='center',
        fontsize=8,
        color='black',
        alpha=0.7
    )

    plt.xlabel('height (cm)')
    plt.ylabel('Weight (kg)')
    plt.title('Buble Chart: Height vs Weight with Age as Bubble Size and Student Names')

    # Add Colorbar
    cbar = plt.colorbar(scatter)
    cbar.set_label('Age (years)')

    plt.show()
```
