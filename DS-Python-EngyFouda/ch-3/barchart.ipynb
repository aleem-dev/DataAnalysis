# Barchart
Barchart is used to show comparision side by side

Here we are going to use NYPD dataset to know which borough has highest crime.  We will use three columns out of the massive dataset, borough, latitude and longitude

```python
import pandas as pd
import matplotlib.pyplot as plt

#import data
#the second approach is faster and less memory consuming always use when working with large files
# df = pd.read_csv('./NYPD_Complaint_Data_Current__Year_To_Date_CP.csv')[['BORO_NM']['Latitude']['Longitude']]
df = pd.read_csv('./NYPD_Complaint_Data_Current__Year_To_Date_CP.csv',usecols=['BORO_NM','Latitude','Longitude'])

#create a table with number of Borough vs crime
rows = df['BORO_NM'].value_counts() # plt.bar(rows.index, rows.values)

# series to dataframe a 2 column table
dfc = pd.DataFrame({'Boroughs':rows.index, 'Crimes':rows.values})

# plot bar chart
plt.figure(figsize=(10,6))
plt.bar(dfc['Boroughs'],dfc['Crimes'].sort_values(ascending=True)) # we can use ascending False to make descending

plt.xlabel('Borough')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes in Different Boroughs')

plt.show()

```
