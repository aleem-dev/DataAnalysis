# Barchart
Barchart is used to show comparision side by side

Here we are going to use NYPD dataset to know which borough has highest crime.  We will use three columns out of the massive dataset, borough, latitude and longitude

```python
import pandas as pd
import matplotlib.pyplot as plt

#import data
#the second approach is faster and less memory consuming always use when working with large files
# df = pd.read_csv('./NYPD_Complaint_Data_Current__Year_To_Date_CP.csv')[['BORO_NM']['Latitude']['Longitude']]
df = pd.read_csv('./NYPD_Complaint_Data_Current__Year_To_Date_CP.csv',usecols=['BORO_NM','Latitude','Longitude'])

#create a table with number of Borough vs crime
rows = df['BORO_NM'].value_counts() # plt.bar(rows.index, rows.values)

# series to dataframe a 2 column table
dfc = pd.DataFrame({'Boroughs':rows.index, 'Crimes':rows.values})

# plot bar chart
plt.figure(figsize=(10,6))
plt.bar(dfc['Boroughs'],dfc['Crimes'].sort_values(ascending=True)) # we can use ascending False to make descending

plt.xlabel('Borough')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes in Different Boroughs')

plt.show()

# we have created dataframe by selecting only 3 columns,
# then we created data series by using .value_counts() 
# it gives counts for each borough name, 
# then we created 2 column table from serias borough and crimes, 
# then we ploted the bar chart.  in barchart on y-axis,
#  we used sort_value(ascending=True) to sort the graph

```

## Sorting



```python
Borough = ['BRONX','BROOKLYN','MANHATTAN','QUEENS','STATEN ISLAND']

crimes = [50153, 67489, 56691, 44137, 10285]

# Sorting
sorted_data = sorted(zip(Borough, crimes), key = lambda x: x[1], reverse = True)

# Extract sorted data
sorted_Borough, sorted_crimes = zip(*sorted_data)

print(f"sorted_Borough: {sorted_Borough}")
print(f"softed crimes: {sorted_crimes}")
```

# Create a Histrogram Using a Bar Chart

We can create histrogram using barchart.  We can also use the string value, if they are in numaric form.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Following is our data
data = {
    'Candidate':['Obama', 'Romney','Johnson','Stein','Paul','Anderson','Reed'],
    'Winning Percentage':[56.27,40.98,1.31,1.14,0.29,0.01,0],
    'colors': ['blue','red','silver','green','grey','grey','grey']
}

df = pd.DataFrame(data)
df.head()

#plot histrogram using barchart
plt.figure(figsize=(10,6))
bars = plt.bar(df['Candidate'],df['Winning Percentage'],color=df['colors'])

# add data labels
for bar, label in zip(bars, df['Winning Percentage']):
    x = bar.get_x() + bar.get_width() / 2
    y = bar.get_height() + 0.5
    plt.text(x, y, f'{label:.2f}%', ha='center', va='bottom', color='black', fontweight='bold')

plt.xlabel('Candidates')
plt.ylabel('Winning Percentages (%)')
plt.title('Winning Percentages of Presidential Candidates')

# plt.show()
```
