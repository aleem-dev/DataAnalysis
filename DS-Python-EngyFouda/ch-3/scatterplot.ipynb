```python
# prepare the chart
import pandas as pd
import matplotlib.pyplot as plt

#fetch data
df = pd.read_csv('./City_of_Seattle_Wages__Comparison_by_Gender_-_Average_Hourly_Wage_by_Age.csv')
#df.info()
#df.head()
#df.describe()

#prepare plot
plt.figure(figsize = (12,6))
df1 = df['AGE RANGE']
df2 = df['Total Average of HOURLY RATE']
plt.scatter(df1,df2)
plt.xlabel('AGE RANGE')
plt.ylabel('Total Average of HOURLY RATE')
plt.title('Interdependence between Age Range and Total Average of HOURLY RATE')
```
