## Correlation Analysis
Before explaining the correlation concepts, we need to undestand difference between the causation and the correlation. CAusation is when one thing causes another thing. While correlation is when two or more things apear to be related. Correlations do not always mean causation.

For example, trave causes buying suitacases and tickets, but suitcase sales and tickets are correlated.

Correlation serves as a statistical method to depect the connection between numerical variables.  This connection is articulated throungh the computation of correlation coefficients, which range from -1 to 1.  The correlation anlaysis task furnishes both graphs and statistics to explore the associations among variables.

Let us inspect a couple fo examples.  The first one is finding if there is a correlation between the lenght and weight of cars and their horsepower.

```python
#correlation between car length, weight and horsepower
import pandas as pd
import matplotlib.pyplot as plt

#import data from csv
cars_df = pd.read_excel('./CARS.xlsx')
# cars_df.info()

#create new data frame
corr_columns = cars_df[['Horsepower','Weight','Length']]
corr_columns.info()

corr_Weight = corr_columns['Weight'].corr(corr_columns['Horsepower'])
print(f'Horsepower Corralation with Weight: {corr_Weight}')

corr_Length = corr_columns['Length'].corr(corr_columns['Horsepower'])
print(f'Horsepower Corralation with Length: {corr_Length}')

#plot scattered plot
plt.figure(figsize=(8,6))
plt.scatter(corr_columns['Weight'],corr_columns['Horsepower'])
plt.title('Correlation between Weight and Power')
plt.show()
plt.figure(figsize=(8,6))
plt.scatter(corr_columns['Length'],corr_columns['Horsepower'])
plt.title('Correlation between Lenght and Power')
plt.show()
```
